```python
df3=df2.sort_values(by='neighbourhood')
#print df3['neighbourhood']

dictionary2={}
dictionary3={}
for ident in df3.index.values:
    if df3['pre 2014 reviews'][ident]==0:
        df3=df3.drop(ident)
    elif df3['post 2014 reviews'][ident]==0:
        df3=df3.drop(ident)

for ident in df3.index.values:
    if df3['neighbourhood'][ident] in dictionary2:
        #dictionary3[df3['neighbourhood'][ident]].append(df3['price'][ident])
        dictionary2[df3['neighbourhood'][ident]].append((df3['pre 2014 reviews'][ident],df3['post 2014 reviews'][ident]))
        
    else:
        #dictionary3[df3['neighbourhood'][ident]]=[]
        #dictionary3[df3['neighbourhood'][ident]].append(df3['price'][ident])

        dictionary2[df3['neighbourhood'][ident]]=[]
        dictionary2[df3['neighbourhood'][ident]].append((df3['pre 2014 reviews'][ident],df3['post 2014 reviews'][ident]))

        


for ident in df2.index.values:
    if df2['neighbourhood'][ident] not in dictionary3:
        dictionary3[df2['neighbourhood'][ident]]=[]
    if df2['neighbourhood'][ident] in dictionary2:
        dictionary3[df2['neighbourhood'][ident]].append(df2['price'][ident])
    

postReviews={}
preReviews={}
priceAvgs={}


#print len(dictionary3['neighbourhood']['78749'])
for ident in df3.index.values:
    priceAvgs[df3['neighbourhood'][ident]]=sum(x for x in dictionary3[df3['neighbourhood'][ident]])/len(dictionary3[df3['neighbourhood'][ident]])
    postReviews[df3['neighbourhood'][ident]]=sum(x[1] for x in dictionary2[df3['neighbourhood'][ident]])
    preReviews[df3['neighbourhood'][ident]]=sum(x[0] for x in dictionary2[df3['neighbourhood'][ident]])


print priceAvgs
#print postReviews
#print preReviews
```

```python
zipCodesPercentChange={}
priceAvgFinal={}
for val in postReviews:
    zipCodesPercentChange[val]=(postReviews[val]-preReviews[val])*100.0/abs(preReviews[val])
    priceAvgFinal[val]=priceAvgs[val]
    
print zipCodesPercentChange

dfZipGrowth = pd.DataFrame(index=zipCodesPercentChange.keys(), columns=['Percent Change 2014-2019','Prices'])
#dfZipGrowth['Zip Code']=zipCodesPercentChange.keys()
dfZipGrowth['Percent Change 2014-2019']=zipCodesPercentChange.values()
dfZipGrowth['Prices']=priceAvgs.values()
dfZipGrowth=dfZipGrowth.sort_values(by=['Percent Change 2014-2019'])
dfZipGrowth.to_csv('zip code percent changes.csv')
```
